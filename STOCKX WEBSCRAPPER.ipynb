```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for points in soup.find_all('h1'):
    name = str(points.text)
```

```python
desr=soup.find_all(attrs={"class":"product-description description-expanded"})
try:
    if not desr:
        desr=soup.find_all(attrs={"class":"product-description description-condensed"})
        desr1=str(desr)
        descript1=desr1.split(">")[2][:-4]
        descript2=desr1.split(">")[4][:-3]
        m=descript1 + descript2
        description=m.replace('StockX','FlipX')
        print(description)
    else:
        desr1=str(desr)
        descript1=desr1.split(">")[2][:-4]
        descript2=desr1.split(">")[4][:-3]
        m=descript1 + descript2
        description=m.replace('StockX','FlipX')
        print(description)
except:
    print("Error Occured")
```

```python
style=soup.find_all(attrs={"data-testid":"product-detail-style"})
style1=str(style)
style_final=style1.split(">")[1][1:-7]
style_final
```

```python
color=soup.find_all(attrs={"data-testid":"product-detail-colorway"})
color1=str(color)
color_final=color1.split(">")[1][1:-7]
color_final
```

```python
retail=soup.find_all(attrs={"data-testid":"product-detail-retail price"})
retail1=str(retail)
retail_final=retail1.split(">")[1][2:-7]
retail_price=int(retail_final)
retail_price_rs=retail_price*75
```

```python
date=soup.find_all(attrs={"data-testid":"product-detail-release date"})
date1=str(date)
date_final=date1.split(">")[1][1:-7]
date_final
```

```python
s = date_final
def swap_(s):
    a = s.split('/')
    a[0], a[1] = a[1], a[0]
    return '/'.join(a)

date_final_indian=swap_(s)
date_final_indian
```

```python
image_link=soup.find_all(attrs={"data-testid":"product-detail-image"})
image_link1=str(image_link)
imgURL=image_link1.split(",")[6][:-12]
imgURL
```

```python
# Importing required libraries
import urllib.request
filenam=name.replace(" ","_")
filename=filenam +".jpg"
urllib.request.urlretrieve(imgURL,filename)
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
def print_properly(string,var):
    print(color.BOLD+string+color.END+var)
print_properly("Name- ",name)
print_properly("Style- ",style_final)
print_properly("Colorway- ",color_final)
print_properly("Retail price(₹)- ",str(retail_price_rs))
print_properly("Release Date- ",date_final)
print_properly("Description- ",description)

from IPython.display import Image
Image(filename,width=300,height=300)
```
